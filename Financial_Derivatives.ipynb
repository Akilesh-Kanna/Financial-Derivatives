```python
#Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# For making attractive and informative statistical graph
sns.set_style('darkgrid')
#plt.style.use('seaborn-darkgrid')
# Define a function put_payoff that calculates the payoff from buying a put option
def put_payoff (sT, strike_price, premium):

    pnl = np.where(sT<strike_price, strike_price - sT, 0)

    return pnl - premium
# We will define the spot price, the strike price, premium, and a range of possible values for the Infosys
#stock price at expiry
# Infosys stock price
spot_price = 900
# Put strike price and cost
strike_price = 900
premium = 20
# Stock price range at the expiration of the put
# We have defined range for the stock price at expiry as +/- 10% from spot price
# Syntax: numpy.arange(start price, stop price)

sT = np.arange(0.9*spot_price,1.1*spot_price)
# Put option buyer payoff
payoff_long_put = put_payoff(sT, strike_price, premium)

# Plot the graph
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(sT,payoff_long_put,label='Put option buyer payoff')
plt.xlabel('Infosys Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
strike_price = 100
premium = 10
stock_prices = np.arange(50, 151, 0.1)  # Equally spaced stock prices from 50 to 150

# Calculate payoff
payoff = np.maximum(0, stock_prices - strike_price) - premium

# Plot the payoff diagram
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, payoff, label='Call Option Payoff')
plt.xlabel('Stock Price')
plt.ylabel('Profit/Loss')
plt.title('Call Option Payoff Diagram')
plt.grid(True)
plt.axhline(y=0, color='black', linestyle='--')  # Add a horizontal line at y=0
plt.axvline(x=strike_price, color='gray', linestyle='--') # Add a vertical line at strike price
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
strike_price = 100
call_premium = 5
put_premium = 4
stock_prices = np.arange(50, 151, 1)  # Stock prices from 50 to 150

# Calculate call payoff
call_payoff = np.maximum(0, stock_prices - strike_price) - call_premium

# Calculate put payoff
put_payoff = np.maximum(0, strike_price - stock_prices) - put_premium

# Calculate straddle payoff
straddle_payoff = call_payoff + put_payoff

# Plot the payoff diagram
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, straddle_payoff, label='Straddle Payoff')
plt.xlabel('Stock Price')
plt.ylabel('Profit/Loss')
plt.title('Straddle Option Payoff Diagram (Strike Price = 100)')
plt.grid(True)
plt.axhline(y=0, color='black', linestyle='--')  # Add a horizontal line at y=0
plt.axvline(x=strike_price, color='gray', linestyle='--') # Add a vertical line at strike price
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
stock_price = 100
call1_strike = 95
call1_premium = 7
call2_strike = 105
call2_premium = 3
stock_prices = np.arange(70, 131, 1)  # Stock prices from 70 to 130

# Calculate payoff for the long call with strike price 95
call1_payoff = np.maximum(0, stock_prices - call1_strike) - call1_premium

# Calculate payoff for the short call with strike price 105
call2_payoff = -(np.maximum(0, stock_prices - call2_strike) - call2_premium)

# Calculate bull call spread payoff
bull_call_spread_payoff = call1_payoff + call2_payoff

# Plot the payoff diagram
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, bull_call_spread_payoff, label='Bull Call Spread Payoff')
plt.xlabel('Stock Price')
plt.ylabel('Profit/Loss')
plt.title('Bull Call Spread Option Payoff Diagram')
plt.grid(True)
plt.axhline(y=0, color='black', linestyle='--')  # Add a horizontal line at y=0
plt.axvline(x=call1_strike, color='gray', linestyle='--') # Add a vertical line at call1 strike price
plt.axvline(x=call2_strike, color='gray', linestyle='--')  # Add a vertical line at call2 strike price
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
stock_price = 60
call_strike = 65
put_strike = 55
call_premium = 1.75
put_premium = 2.625
stock_prices = np.arange(40, 81, 1)  # Stock prices from 40 to 80

# Calculate payoff for the short call
short_call_payoff = -(np.maximum(0, stock_prices - call_strike) - call_premium)

# Calculate payoff for the short put
short_put_payoff = -(np.maximum(0, put_strike - stock_prices) - put_premium)

# Calculate short strangle payoff
short_strangle_payoff = short_call_payoff + short_put_payoff

# Plot the payoff diagram
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, short_strangle_payoff, label='Short Strangle Payoff')
plt.xlabel('Stock Price')
plt.ylabel('Profit/Loss')
plt.title('Short Strangle Option Payoff Diagram')
plt.grid(True)
plt.axhline(y=0, color='black', linestyle='--')  # Add a horizontal line at y=0
plt.axvline(x=put_strike, color='gray', linestyle='--')  # Add a vertical line at put strike price
plt.axvline(x=call_strike, color='gray', linestyle='--')  # Add a vertical line at call strike price
plt.legend()
plt.show()

# Payoff calculations table
data = []
for i, price in enumerate(stock_prices):
    call_payoff = -(max(0, price - call_strike) - call_premium)
    put_payoff = -(max(0, put_strike - price) - put_premium)
    total_payoff = call_payoff + put_payoff
    data.append([price, call_payoff, put_payoff, total_payoff])

# Create a pandas DataFrame for better visualization
import pandas as pd
df = pd.DataFrame(data, columns=['Stock Price', 'Short Call Payoff', 'Short Put Payoff', 'Total Payoff'])
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define stock price range
stock_prices = np.linspace(15, 35, 200)  # Stock prices from $15 to $35

# Parameters for bought and sold calls
strike_bought = 19.12
strike_sold = 27.3175
premium_bought = 5.38
premium_sold = 2.96

# Payoffs for individual options
payoff_bought = np.maximum(stock_prices - strike_bought, 0) - premium_bought
payoff_sold = premium_sold - np.maximum(stock_prices - strike_sold, 0)

# Combined call spread payoff
payoff_call_spread = payoff_bought + payoff_sold

# Plot the payoff diagrams
plt.figure(figsize=(12, 8))

# Bought call option
plt.plot(stock_prices, payoff_bought, label=f"Bought Call ($19.12 Strike, $5.38 Premium)", color="blue", linestyle="--")

# Sold call option
plt.plot(stock_prices, payoff_sold, label=f"Sold Call ($27.3175 Strike, $2.96 Premium)", color="orange", linestyle="--")

# Combined call spread
plt.plot(stock_prices, payoff_call_spread, label="Net Payoff (Call Spread)", color="green", linewidth=2)

# Add markers for strike prices
plt.axvline(x=strike_bought, color="blue", linestyle=":", label=f"Bought Strike Price: ${strike_bought}")
plt.axvline(x=strike_sold, color="orange", linestyle=":", label=f"Sold Strike Price: ${strike_sold}")

# Add labels, legend, and grid
plt.axhline(y=0, color="black", linestyle="-")
plt.title("Payoff Diagram: Bought and Sold Calls with Net Call Spread", fontsize=14)
plt.xlabel("Stock Price at Expiry ($)", fontsize=12)
plt.ylabel("Payoff ($)", fontsize=12)
plt.legend(fontsize=10)
plt.grid(True)
plt.show()
```
